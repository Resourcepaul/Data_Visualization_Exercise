```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('/content/drive/MyDrive/Bootcamp/04 Week 4: Explanatory Viz/Income Inequality.xlsx')
```

```python
df
```

```python
group_df=df[df['G7']=='Y'].groupby(['COUNTRY'])['VALUE'].mean().reset_index()
```

```python
group_df.head()
```

```python
sns.set(rc={'figure.figsize':(20,10)})

sns.barplot(data=group_df,y=group_df.COUNTRY,x='VALUE',order=group_df.sort_values('VALUE',ascending=False).COUNTRY);
```
